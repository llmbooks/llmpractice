```python
import matplotlib.pyplot as plt  
import numpy as np  
import matplotlib.ticker as ticker 
from matplotlib.gridspec import GridSpec


N = np.logspace(5, 9, num=10)  # Creates 100 points between 10^5 and 10^9  
L1 = (N / (8.8 * 10**13)) ** -0.076

D = np.logspace(7, 9, num=10)
L2 = (D / (5.4 * 10**13)) ** -0.095

C = np.logspace(-9, 1, num=10)
L3 = (C / (3.1 * 10**8)) ** -0.050

  
# Creating the plot  
fig = plt.figure(figsize=(12, 3), tight_layout=True)
gs = GridSpec(1, 3)
X = [N, D, C]
Y = [L1, L2, L3]
labels = [
    '$L = (N/8.8 \cdot 10^{13})^{-0.076}$',
    '$L = (D/5.4 \cdot 10^{13})^{-0.095}$',
    '$L = (C_{\min}/3.1 \cdot 10^8)^{-0.050}$',
]
xlabels = ['$N$', '$D$', '$C_{\min}$']
titles = ['', '', '']
ymax = [5.5, 5, 8]
ystep = [0.5, 0.5, 1]

# Set y-axis to display plain numbers (without scientific notation)  
def format_func(value, tick_number):  
    # Return the value without any formatting  
    return f'{value}'  

for i in range(3):
    ax = fig.add_subplot(gs[0,i])
    ax.loglog(X[i], Y[i], linestyle='-', color='blue', label=labels[i])  
  
    # Adding title and labels  
    ax.title.set_text(titles[i])
    ax.title.set_fontname('SimHei') 
    ax.set_xlabel(xlabels[i])  
    if i == 0: ax.set_ylabel('L')

    # Adding the legend  
    ax.legend()  
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_func))
    ax.set_yticks(np.arange(2, ymax[i], ystep[i]))

# Displaying the plot  
#plt.show() 
plt.savefig('power_law.png', dpi=600, bbox_inches='tight')
```
