```python
import numpy as np
import matplotlib.pyplot as plt

def get_positional_encoding(max_seq_len, d_model):
    position = np.arange(max_seq_len)[:, np.newaxis]
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
    pos_enc = np.zeros((max_seq_len, d_model))
    pos_enc[:, 0::2] = np.sin(position * div_term)
    pos_enc[:, 1::2] = np.cos(position * div_term)
    return pos_enc

max_seq_len = 50
d_model = 16

pos_enc = get_positional_encoding(max_seq_len, d_model)

plt.figure(figsize=(10,5))
plt.pcolormesh(pos_enc, cmap='coolwarm')
plt.xlabel('特征维度', font='SimHei', fontsize=14)
plt.xlim((0, d_model))
plt.ylim((max_seq_len,0))
plt.ylabel('词元位置', font='SimHei', fontsize=14)
plt.colorbar()
plt.savefig('positional_encoding.png', dpi=600, bbox_inches='tight')
```
